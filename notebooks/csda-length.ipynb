# Evaluating CSDA/Length

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CSDA = "/Volumes/easystore/proton-kaon/docs/selected_kaon_csda.csv"
MODEL = "/Volumes/easystore/proton-kaon/docs/selected_kaon_model.csv"
MASS = "/Volumes/easystore/proton-deuteron/csv/picky+match.csv"
```

```python
csda = pd.read_csv(CSDA)
model = pd.read_csv(MODEL).drop(columns='Unnamed: 0')
mass = pd.read_csv(MASS)
```

```python
csda = csda.merge(mass, how='inner', on=['run', 'subrun', 'event'])
csda
```

```python
plt.hist(csda['beamline_mass'].values, bins='auto')
plt.xlabel('Beamline Mass [MeV/c^2]')
plt.ylabel('Count')
plt.title('Kaons Selected Using CSDA/Length')
plt.show()
```
